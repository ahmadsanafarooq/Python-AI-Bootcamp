```python
#libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
#scatter plot
sns.scatterplot(data=df, x='age', y='fare', hue='class')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='fare', hue='survived')
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='survived',
    palette='viridis',
    sizes=(20, 200),
    alpha=0.5
)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='survived',
    palette='dark',     
    sizes=(20, 200),
    alpha=0.5
)
plt.title("Age vs Fare by Survival (Dark Colors)")
plt.show()
```

```python
# Set dark background
plt.style.use('dark_background')

# Optional: Choose a deep or dark Seaborn palette
sns.set_palette('deep')  # or try 'dark'

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='survived',
    palette='deep',     # or 'dark'
    sizes=(20, 200),
    alpha=0.6
)

plt.title("Age vs Fare by Survival", color='white')
plt.xlabel("Age", color='white')
plt.ylabel("Fare", color='white')
plt.legend(title='Survived')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
```

```python
palette = {0: '#1f1f1f', 1: '#003f5c'} 

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='age',
    y='fare',
    hue='survived',
    palette=palette,
    sizes=(20, 200),
    alpha=0.6
)
plt.title("Age vs Fare by Survival (Custom Dark Palette)")
plt.show()
```
